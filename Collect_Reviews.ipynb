```python
from selenium import webdriver
import time
import pandas as pd
import os
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.capterra.fr/reviews/120550/asana-pm"
```

```python
options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('--disable-infobars')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
options.add_argument('--remote-debugging-port=9222')

driver = webdriver.Chrome(options=options)
driver.implicitly_wait(10)
driver.get(url)
```

```python
reviews = driver.find_elements(By.CSS_SELECTOR, "div.i18n-translation_container")
n = len(reviews)
n
```

```python
base_url = "https://www.capterra.fr/reviews/120550/asana-pm?page="

all_reviews = []

for page in range(1, 2):
    page_url = base_url + str(page)
    driver.get(page_url)  # Charger la page
    review_containers = driver.find_elements(By.CLASS_NAME, "review-card")
    print(page)
    for review_container in review_containers:
        try:
            client_name = review_container.find_element(By.CLASS_NAME, "fw-bold").text
            client_job = review_container.find_element(By.CLASS_NAME, "text-ash").text
            review_name = review_container.find_element(By.TAG_NAME, "h3").text
            date_review = review_container.find_element(By.XPATH, ".//span[@class='ms-2']").text
            période_utilisation = review_container.find_element(By.XPATH, ".//div[@class='mb-2']/following-sibling::div[1]").text
            comment = review_container.find_element(By.XPATH, ".//p/span[contains(@class, 'fw-bold')]/following-sibling::span").text
            avantages = review_container.find_element(By.XPATH, ".//div[@class='col-lg-7']/p[contains(@class, 'fw-bold')]/following-sibling::p[1]").text
            inconvénients = review_container.find_element(By.XPATH, ".//div[@class='col-lg-7']/p[contains(@class, 'fw-bold')]/following-sibling::p[3]").text
            score = review_container.find_element(By.XPATH, " .//div[@class='text-ash mb-3']/span[contains(@class, 'mos-star-rating')]/span[@class='d-flex']/span[@class='ms-1']").text

            all_reviews.append({
                "client_name": client_name,
                "client_job": client_job,
                "review_name": review_name,
                "data_review": date_review,
                "période_utilisation": période_utilisation,
                "comment":  comment,
                "avantages": avantages,
                "inconvénients": inconvénients,
                "score": score
            })
        except Exception as e:
            print("information not found for URL:", page_url)
```
